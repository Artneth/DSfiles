#Concept of Standardization (Z-Score Normalization)
Standardization is a technique used to scale numerical variables to have a consistent range and distribution, which is particularly important when features have varying scales in a dataset. It ensures that all variables contribute equally to the model, preventing features with larger magnitudes from dominating the analysis. This also allows easier comparison between variables on different scales.

Formula:
The formula for standardization is: z = (x - μ)/ σ

Where:
x: A data point from the feature.
μ: Mean of the feature.
σ: Standard deviation of the feature.



```python
import numpy as np
```

```python
standardized_data = (numerical_columns - numerical_columns.mean()) / numerical_columns.std()
standardized_data.head()
```

```python
standardized_data.describe()
```

```python
fig, axes = plt.subplots(2, 6, figsize=(20, 8))

axes[0, 0].hist(numerical_columns['Volume'], bins=30, color='blue')
axes[0, 0].set_title('Volume - Original')
axes[1, 0].hist(standardized_data['Volume'], bins=30, color='green')
axes[1, 0].set_title('Volume - Standardized')

axes[0, 1].hist(numerical_columns['Avg Price'], bins=30, color='blue')
axes[0, 1].set_title('Avg Price - Original')
axes[1, 1].hist(standardized_data['Avg Price'], bins=30, color='green')
axes[1, 1].set_title('Avg Price - Standardized')

axes[0, 2].hist(numerical_columns['Total Sales Value'], bins=30, color='blue')
axes[0, 2].set_title('Total Sales Value - Original')
axes[1, 2].hist(standardized_data['Total Sales Value'], bins=30, color='green')
axes[1, 2].set_title('Total Sales Value - Standardized')

axes[0, 3].hist(numerical_columns['Discount Rate (%)'], bins=30, color='blue')
axes[0, 3].set_title('Discount Rate (%) - Original')
axes[1, 3].hist(standardized_data['Discount Rate (%)'], bins=30, color='green')
axes[1, 3].set_title('Discount Rate (%) - Standardized')

axes[0, 4].hist(numerical_columns['Discount Amount'], bins=30, color='blue')
axes[0, 4].set_title('Discount Amount - Original')
axes[1, 4].hist(standardized_data['Discount Amount'], bins=30, color='green')
axes[1, 4].set_title('Discount Amount - Standardized')

axes[0, 5].hist(numerical_columns['Net Sales Value'], bins=30, color='blue')
axes[0, 5].set_title('Net Sales Value - Original')
axes[1, 5].hist(standardized_data['Net Sales Value'], bins=30, color='green')
axes[1, 5].set_title('Net Sales Value - Standardized')

plt.tight_layout()
plt.show()
```

# Conversion of Categorical Data into Dummy Variables
Converting categorical data into dummy variables (one-hot encoding) is essential because most machine learning models work with numerical data, not categorical. One-hot encoding transforms categorical variables into a format that can be provided to ML algorithms by creating binary (0/1) columns for each category, ensuring:

* No Ordinal Misinterpretation: Prevents algorithms from assuming a natural order in categories.
* Model Compatibility: Allows algorithms like regression or neural networks to process categorical features.
* Preserves Information: Encodes all categories without merging or losing distinctions.

This process ensures the model understands and learns from categorical data
effectively.

```python
categorical_cols = data.select_dtypes(include=['object']).columns

data_encoded = pd.get_dummies(data, columns=categorical_cols)
data_encoded.iloc[10:20, :20]
```
